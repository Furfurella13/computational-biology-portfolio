```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def gabor_function(x, gamma=1, omega=1):
    return np.exp(-(x**2/2*gamma**2))*np.cos(2*np.pi*omega*x) #функция Габора

def gabor_function_der(x, gamma=1, omega=1):
    return -np.e**(-gamma**2*x**2/2)*(2*np.pi*omega*np.sin(2*np.pi*omega*x) + gamma**2*x*np.cos(2*np.pi*omega*x)) #производная функции Габора
```

```python
x = np.linspace(-5, 5, 1000)
y = np.apply_along_axis(gabor_function, axis=0, arr=x)
```

```python
y_der = np.apply_along_axis(gabor_function_der, axis=0, arr=x)
```

```python
plt.plot(x, y_der)
plt.plot(x, y); #график функции Габора и его производной
```

```python
plt.figure(figsize=(12,7))

lr = 0.001  #выставляем лернинг рейт

start_point = np.random.uniform(-3, 3) #стартовая точка по оси х
start_value = gabor_function(start_point) # рандомное стартовое значение

plt.scatter(x=[start_point], y=[start_value], marker="X", s=400, c="green") #обозначение начала спуска


plt.plot(x, y);

for _ in range(200):
    derrivative = gabor_function_der(start_point) #производная текущей точки
    start_point = start_point - lr*derrivative #делаем градиентный шаг
    plt.scatter(x=[start_point], y=[gabor_function(start_point)])

plt.scatter(x=[start_point], y=[gabor_function(start_point)], marker="*", s=400, c="red")  #обозначение конца спуска

plt.show()
```

```python
plt.figure(figsize=(12,7))

lr = 0.001
momentum = 0.1 #добавляем моментум

start_point = np.random.uniform(-3, 3)
start_value = gabor_function(start_point)

delta_w = 0
plt.scatter(x=[start_point], y=[start_value], marker="X", s=400, c="green")


plt.plot(x, y);

for _ in range(100):
    derrivative = gabor_function_der(start_point)
    delta_w = -lr*derrivative + momentum*delta_w
    start_point = start_point + delta_w
    plt.scatter(x=[start_point], y=[gabor_function(start_point)])

plt.scatter(x=[start_point], y=[gabor_function(start_point)], marker="*", s=400, c="red")

plt.show()
```
